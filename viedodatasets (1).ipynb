```python
import cv2
import os

def extract_frames(video_path, output_dir, frame_rate=1):
    """
    Extract frames from a video at a given frame rate.

    Args:
    - video_path (str): Path to the video file.
    - output_dir (str): Directory where the frames will be saved.
    - frame_rate (int): Number of frames to extract per second (1 frame/sec).
    """
    # Open the video file
    video = cv2.VideoCapture(video_path)

    if not video.isOpened():
        print(f"Error: Could not open video {video_path}")
        return
    
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    frame_count = 0
    success, frame = video.read()

    if not success:
        print(f"Error: Failed to read the first frame of {video_path}")
        return

    while success:
        # Save frames at intervals based on frame_rate
        if frame_count % frame_rate == 0:
            frame_name = f"{output_dir}/frame_{frame_count}.jpg"
            cv2.imwrite(frame_name, frame)
            print(f"Extracted and saved frame {frame_count} for video: {video_path}")
        
        # Read next frame
        success, frame = video.read()
        frame_count += 1

    video.release()
    print(f"Frames extracted and saved to {output_dir} for video: {video_path}")


# List of 9 video file paths
video_paths = [
    "C:/Users/sivad/OneDrive/Desktop/viedos/video 1.mp4",
    "C:/Users/sivad/OneDrive/Desktop/viedos/video 2.mp4",
    "C:/Users/sivad/OneDrive/Desktop/viedos/video 3.mp4",
    "C:/Users/sivad/OneDrive/Desktop/viedos/video 4.mp4",
    "C:/Users/sivad/OneDrive/Desktop/viedos/video 5.mp4",
    "C:/Users/sivad/OneDrive/Desktop/viedos/video 6.mp4",
    "C:/Users/sivad/OneDrive/Desktop/viedos/video 7.mp4",
    "C:/Users/sivad/OneDrive/Desktop/viedos/video 8.mp4",
    "C:/Users/sivad/OneDrive/Desktop/viedos/video 9.mp4"
]

# Loop over each video and extract frames
for i, video_path in enumerate(video_paths):
    output_dir = f"output_frames_directory_video_{i+1}"  # Unique output directory for each video
    extract_frames(video_path, output_dir, frame_rate=10)  # Extract 1 frame every 10 seconds

```

```python
import cv2
import os

def preprocess_frame(frame_path, target_size=(224, 224), grayscale=False):
    """
    Preprocess a single image frame.
    
    Args:
    - frame_path (str): Path to the image frame.
    - target_size (tuple): Desired size for the image.
    - grayscale (bool): Whether to convert to grayscale.
    
    Returns:
    - preprocessed_frame (numpy array): Preprocessed image frame.
    """
    # Read the frame
    frame = cv2.imread(frame_path)
    
    if frame is None:
        raise ValueError(f"Error reading the frame {frame_path}. Make sure it's a valid image.")
    
    # Convert to grayscale if needed
    if grayscale:
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Resize the image
    frame = cv2.resize(frame, target_size)

    # Normalize the pixel values (scale between 0 and 1)
    frame = frame / 255.0

    return frame

# Example: Preprocess a frame from video 1
frame_path = "output_frames_directory_video_1/frame_0.jpg"
preprocessed_frame = preprocess_frame(frame_path, target_size=(224, 224), grayscale=True)
print(f"Preprocessed frame shape: {preprocessed_frame.shape}")
```

```python
import os
import numpy as np
from keras.applications import ResNet50
from keras.applications.resnet50 import preprocess_input
from keras.preprocessing import image

# Load ResNet50 model (without the top classification layer)
model = ResNet50(weights='imagenet', include_top=False)

# Function to extract features from a single image frame
def extract_features_from_frame(frame_path, target_size=(224, 224)):
    """
    Extract features from a single image frame using a pre-trained ResNet50 model.
    
    Args:
    - frame_path (str): Path to the image frame.
    - target_size (tuple): Desired size for the image.
    
    Returns:
    - features (numpy array): Extracted features for the frame.
    """
    # Load and preprocess the frame
    img = image.load_img(frame_path, target_size=target_size)
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)

    # Extract features using ResNet50
    features = model.predict(x)
    
    # Flatten the features
    features = features.flatten()

    return features

# Lists to store the extracted features and labels
frame_features = []
labels = []

# Function to extract features from all frames in a directory
def extract_features_from_all_frames(video_dir, label, target_size=(224, 224)):
    """
    Extract features from all frames in a directory and label them.

    Args:
    - video_dir (str): Directory where the frames are stored.
    - label (int): Label for the frames (0 = non-fraud, 1 = fraud).
    - target_size (tuple): Desired size for the images.
    """
    for frame_file in os.listdir(video_dir):
        frame_path = os.path.join(video_dir, frame_file)
        if frame_file.endswith(".jpg"):  # Ensure we only process image files
            features = extract_features_from_frame(frame_path, target_size)
            frame_features.append(features)
            labels.append(label)

# Example: Process frames from video 1 (adjust label based on fraud/non-fraud)
extract_features_from_all_frames("output_frames_directory_video_1", label=0)  # Label for video 1

# Process frames for other videos (modify labels as needed)
extract_features_from_all_frames("output_frames_directory_video_2", label=1)  # Label for video 2
extract_features_from_all_frames("output_frames_directory_video_3", label=0)  # Label for video 3
# Repeat for other videos...

# Convert to numpy arrays
X = np.array(frame_features)
y = np.array(labels)

print(f"Features names: {X.names}")
print(f"Labels shape: {y.shape}")

```
